i wanted to check how does the gross looks for the movies

```python
features_start_at = list(df1.columns).index('duration')
feature_names = df1.columns[features_start_at:]
feature_names

#i set this cause in the below snippet i can even generate plots for all columns at once
```

```python
fig, axs = plt.subplots(figsize=(12, 6), dpi=400)
#axs = axs.flatten() 
for i,n in enumerate(feature_names):
    ax = axs
    ax.bar(
        df1['movie_title'], df1['gross'])
    
    #ax.set_ylabel(i)
    ax.set_ylabel('gross')
    ax.set_xticklabels(df1.movie_title.values, rotation=90)
    #ax.set_title(df1['director_name'].values[i])

    
plt.tight_layout()
plt.show()
#fig.savefig('name.jpg', dpi=400)
```
